```python
from lib.pipeline import Pipeline
from lib.disable_logger import DisableLogger
ls = []
with DisableLogger():
    for trial in range(3):
        pipeline = Pipeline(
            model='cnn',
            dataset='atpbind',
            gpus=[1],
            model_kwargs={
                "input_dim": 21,
                "hidden_dims": [1024, 1024],
                "kernel_size": 5,
                "padding": 2,
                "readout": "max",
            }
        )
        for epoch in range(10):
            pipeline.train(num_epoch=1)
            ls.append({"trial": trial, "epoch": epoch, "data": pipeline.evaluate()})
```

```python
from statistics import mean, stdev
auroc = [d["data"]["micro_auroc"] for d in ls]
mcc = [d["data"]["mcc"] for d in ls]

print(f'AUROC: {mean(auroc):.3f} ± {stdev(auroc):.3f}')
print(f'MCC: {mean(mcc):.3f} ± {stdev(mcc):.3f}')
```

```python
from lib.pipeline import Pipeline
from lib.disable_logger import DisableLogger
ls_3 = []
with DisableLogger():
    for trial in range(3):
        pipeline = Pipeline(
            model='cnn',
            dataset='atpbind',
            gpus=[1],
            model_kwargs={
                "input_dim": 21,
                "hidden_dims": [1024, 1024, 1024],
                "kernel_size": 5,
                "padding": 2,
                "readout": "max",
            }
        )
        for epoch in range(5):
            pipeline.train(num_epoch=1)
            ls_3.append({"trial": trial, "epoch": epoch, "data": pipeline.evaluate()})
```

```python
from statistics import mean, stdev
auroc = [d["data"]["micro_auroc"] for d in ls_3 if d["epoch"] == 4]
mcc = [d["data"]["mcc"] for d in ls_3 if d["epoch"] == 4]

print(f'AUROC: {mean(auroc):.3f} ± {stdev(auroc):.3f}')
print(f'MCC: {mean(mcc):.3f} ± {stdev(mcc):.3f}')
```
