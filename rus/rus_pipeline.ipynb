```python
import pandas as pd

df = pd.read_csv('rus_pipeline_v3.csv')
df_stat = df.groupby([
    'param_version'
    ]
).agg({
    'mcc': ['size', 'mean', 'std'],
       'sensitivity': ['mean', 'std'],
    #    'specificity': ['mean', 'std'],
    #    'accuracy': ['mean', 'std'],
       'precision': ['mean', 'std'],
       'valid_mcc': ['mean', 'std'],
       'epoch_count': ['mean', 'std'],
    }
    ).reset_index()

df_stat
```

```python
# 1. Group by 'param_version' and aggregate the metrics without 'size'
df_agg = df.groupby('param_version').agg({
    'mcc': ['mean', 'std'],
    'sensitivity': ['mean', 'std'],
    'precision': ['mean', 'std'],
    'valid_mcc': ['mean', 'std'],
    'epoch_count': ['mean', 'std']
}).reset_index()

# Flatten the multi-level column index
df_agg.columns = ['_'.join(col).strip() for col in df_agg.columns.values]

# Rename the column to remove the appended underscore from 'param_version'
df_agg.rename(columns={'param_version_': 'param_version'}, inplace=True)

# 2. Group by 'param_version' and calculate 'size'
df_size = df.groupby('param_version').size().reset_index(name='size')

# 3. Merge the two dataframes on 'param_version'
df_stat = pd.merge(df_agg, df_size, on='param_version')

df_stat
```
